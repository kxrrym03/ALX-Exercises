# Renaming a column

```python
%%sql
SELECT
Country_name,
Time_period,
Pct_managed_drinking_water_services AS pct_access_to_water
FROM
access_to_basic_services
LIMIT 20;
```

```python
%%sql
SELECT
Country_name,
Time_period,
Pct_managed_drinking_water_services AS pct_access_to_water
FROM
access_to_basic_services
WHERE
Time_period = 2020
LIMIT 20;
```

```python
%%sql
SELECT
Country_name,
Time_period,
Pct_managed_drinking_water_services AS pct_access_to_water
FROM
access_to_basic_services
WHERE
Time_period = 2020
ORDER BY pct_access_to_water
LIMIT 20;
```

# What is the total number of entries in the dataset?

```python
%%sql
SELECT
COUNT(*) AS Number_of_observations
FROM united_nations.Access_to_Basic_Services;
```

# What are the earliest and latest years for which we have data?

```python
%%sql
SELECT
    MIN(Time_period) AS Min_time_period,
    MAX(Time_period) AS Max_time_period
FROM united_nations.Access_to_Basic_Services;
```

# How many countries are included in this dataset?

```python
%%sql

SELECT
    COUNT(DISTINCT Country_name) AS Number_of_countries
FROM united_nations.Access_to_Basic_Services;
```

# What is the average percentage of people who have access to managed drinking water services across all years and all countries included in our dataset?

```python
%%sql

SELECT
    AVG(Pct_managed_drinking_water_services) AS AVG_managed_drinking_water_services
FROM united_nations.Access_to_Basic_Services;
```
